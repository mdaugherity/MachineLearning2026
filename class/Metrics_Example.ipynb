# Metrics Example

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score
```

## Dog Detector
Using the **Dog Detector** from the lecture slides.

There are 20 dogs(+) and 10 cats(-).  A binary classifier correctly predicts 5 dogs and incorrectly predicts 5 cats.




For the dogs we must have 5 true positives and 15 false negatives.  For cats we've got 5 true negatives and 5 false positives.  We can use these numbers to set up predictions

```python
y_true = np.array([1]*20+[0]*10)  # makes 20 1's and 10 0's
y_pred = np.array([1]*5+[0]*15+[0]*5+[1]*5)
print('Actual:  \t', y_true)
print('Predicted:\t', y_pred)
```

```python
confusion_matrix(y_true, y_pred, labels=[1,0])  # show + before - to match slides
```

```python
ConfusionMatrixDisplay.from_predictions(y_true, y_pred, labels=[1,0], cmap='Blues')
plt.show()
```

```python
print(classification_report(y_true, y_pred))
```

Remember that for a binary classifier, we define the overal precision and recall for the **positive** class.  So for this problem the precision is 0.5 and the recall is 0.25

We can also get numbers individually

```python
precision_score(y_true, y_pred)
```

```python
recall_score(y_true, y_pred)
```
